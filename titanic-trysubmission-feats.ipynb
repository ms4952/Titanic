```python
import pandas as pd
import numpy as np
data=pd.read_csv('./titanic/train.csv')
df=pd.DataFrame(data,columns=['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked','Name'])
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
df['Is_Married'] = 0
df['Is_Married'].loc[df['Title'] == 'Mrs'] = 1
df=df.drop(["Title"],axis=1)
df.head()
```

```python
data1=pd.read_csv('./titanic/test.csv')
dft=pd.DataFrame(data1,columns=['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked','Name'])
dft['Title'] = dft['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
dft['Is_Married'] = 0
dft['Is_Married'].loc[dft['Title'] == 'Mrs'] = 1
dft=dft.drop(['Title'],axis=1)
dft.head()
```

```python
df.Cabin=np.where(df.Cabin.isnull(),'N',df['Cabin'])
#df['Cabin']=df['Cabin'].astype('|S')
df.Sex=df.Sex.astype('|S')
df.Embarked=np.where(df.Embarked.isnull(),df['Embarked'].mode(),df['Embarked'])
df.Age=np.where(df.Age.isnull(),df['Age'].mean(),df['Age'])
#df.Age=df.Age.astype(int)
for i in range(0,len(df)):
    df['Cabin'][i]=df['Cabin'][i][0]
df.head()
```

```python
dft.Cabin=np.where(dft.Cabin.isnull(),'N',dft['Cabin'])
#df['Cabin']=df['Cabin'].astype('|S')
dft.Sex=dft.Sex.astype('|S')
dft.Embarked=np.where(dft.Embarked.isnull(),dft['Embarked'].mode(),dft['Embarked'])
dft.Age=np.where(dft.Age.isnull(),dft['Age'].mean(),dft['Age'])
#df.Age=df.Age.astype(int)
for i in range(0,len(dft)):
    dft['Cabin'][i]=dft['Cabin'][i][0]
dft.head()
```

```python
df=pd.get_dummies(df, columns=["Sex","Cabin","Embarked"])


df.head()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
# clf = AdaBoostClassifier(n_estimators=50,learning_rate=1)
# clf.fit(X_train,y_train)
gb=XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=1.0, gamma=1, gpu_id=-1,
              importance_type='gain', interaction_constraints='',
              learning_rate=0.02, max_delta_step=0, max_depth=4,
              min_child_weight=1, monotone_constraints='()',
              n_estimators=600, n_jobs=1, nthread=1, num_parallel_tree=1,
              random_state=0, reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
              silent=True, subsample=0.6, tree_method='exact',
              validate_parameters=1, verbosity=None)
#gb=GradientBoostingClassifier(n_estimators=100, learning_rate = .75, max_features=3, max_depth = 2, random_state = 0)
#gb.fit(X_train,y_train)
gb.fit(X,Y)

y_pred = gb.predict(dft1)
print(y_pred)
```
